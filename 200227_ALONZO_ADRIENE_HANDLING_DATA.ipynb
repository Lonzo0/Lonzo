```python
#problem2
board1 = [
['X','X','O'],
['O','X','O'],
['O','','X'],
]

board2 = [
['X','X','O'],
['O','X','O'],
['','O','X'],
]

board3 = [
['O','X','O'],
['','O','X'],
['X','X','O'],
]

board4 = [
['X','X','X'],
['O','X','O'],
['O','','O'],
]

board5 = [
['X','X','O'],
['O','X','O'],
['X','','O'],
]

board6 = [
['X','X','O'],
['O','X','O'],
['X','',''],
]

    
def tic_tac_toe(board):
    for vertical in [vertical for vertical in zip(*board)]:
        while len(set(vertical)) == 1:
            return vertical[0]

    for horizontal in board:
        while len(set(horizontal)) == 1:
            return horizontal[0]

    if all([diag_updown == 'X' for diag_updown in [board [i][i] for i, v in enumerate(board)]]):
        return 'X'

    elif all([diag_updown == 'O' for diag_updown in [board [i][i] for i, v in enumerate(board)]]):
        return 'O'

    elif all([diag_downup == 'X' for diag_downup in [board [2-i][i] for i, v in enumerate(board)]]):
        return 'X'

    elif all([diag_downup == 'O' for diag_downup in [board [2-i][i] for i, v in enumerate(board)]]):
        return 'O'

    else:
        return None

print(tic_tac_toe(board1))
print(tic_tac_toe(board2))
print(tic_tac_toe(board3))
print(tic_tac_toe(board4))
print(tic_tac_toe(board5))
print(tic_tac_toe(board6))


```

```python
#problem 3

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}


def eta(legs, source, destination):
    
    total_time = 0
    from_ = ''
    schools = list(legs.keys())
    from_list  = [i[0] for i in schools]
    to_list = [i[1] for i in schools]
    tupple = tuple([source, destination])
    
    while from_!= destination:
        if tupple in schools:
            from_ = destination
        
        elif tupple not in schools:
            for j in range(len(legs)):
                if from_list[j] == source:
                    from_ = from_list[j]
                    break
                    
        time = legs[tupple]['travel_time_mins']
        total_time += time
        source = from_
    return total_time

print(eta(legs, 'dlsu', 'upd'))
    
```
